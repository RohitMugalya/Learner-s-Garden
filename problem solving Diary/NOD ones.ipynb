```python
import numpy as np
import pandas as pd
```

```python
def nodones(start, stop):
    count = 0
    for i in range(start, stop+1):
        count += str(i).count('1')
    return count
def nodarr(ar):
    for row in ar:
        count = 0
        for val in row:
            count += str(val).count('1')
        print(f'{row}\t\t -> {count}')

def nodstep(doto, upto):
    for i in range(doto, upto+1, 10):
        count = 0
        for j in range(i):
            count += str(j).count('1')
        print(f'[0, {i-1:>5}]\t -> {count}')

def nodig(n, d):
    c = 0
    d = str(d)
    for i in range(n + 1):
        c += str(i).count(d)
    return c
```
