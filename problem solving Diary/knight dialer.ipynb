```python
phone_pad = {
    1: [8, 6],
    2: [7, 9],
    3: [4, 8],
    4: [3, 9, 0],
    5: [],
    6: [1, 7, 0],
    7: [2, 6],
    8: [1, 3],
    9: [2, 4],
    0: [4, 6],
}
```

```python
def knightDialer(n: int):
    phone_numbers = []
    def dfs(numpad, phone_number):
        if len(phone_number) == n:
            phone_numbers.append(phone_number.copy())
            return
        
        for next_num in phone_pad[numpad]:
            dfs(next_num, phone_number + [next_num])

    for initial_pad in range(10):
        dfs(initial_pad, [initial_pad])
        
    return phone_numbers

def display(phone_numbers):
    print(len(phone_numbers))
    for sno, phone_number in enumerate(phone_numbers, 1):
        print(f"{sno:02}. {phone_number}")
```
