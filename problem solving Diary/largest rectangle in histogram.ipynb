```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use("dark_background")
```

```python
def display(heights, bar_width=1):
    plt.bar(np.arange(len(heights)) - 0.5, heights, width=bar_width, color="brown", edgecolor="white")
    plt.grid()

def generate_heights(start, stop, size):
    return np.random.randint(start, stop + 1, size=size)
```

```python
display([2, 1, 5, 6, 2, 3])
```

```python
h = generate_heights(1, 10, 10)
h
```
