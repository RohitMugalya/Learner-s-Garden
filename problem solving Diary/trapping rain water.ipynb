```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("dark_background")
```

```python
def generate_walls(heights, color="brown", width=1):
    print(heights)
    x = range(len(heights))
    plt.bar(np.array(x) - 0.5, heights, color=color, width=width)
    plt.yticks(x)
    plt.xticks(x)
    plt.grid()

def generate_random(a, b, size=10):
    return np.random.randint(a, b + 1, size=size)
```

```python
heights = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
heights
```

```python
generate_walls(heights)
```
