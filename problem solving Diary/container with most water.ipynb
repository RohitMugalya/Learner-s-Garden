```python
from matplotlib import pyplot as plt
import numpy as np

plt.style.use("dark_background")
```

```python
def generate_walls(heights, water=None, color="brown", width=1):
    print(heights)
    new_lst = []
    for i, height in enumerate(heights):
        new_lst.append(height)
        new_lst.append(0)
    new_lst.pop()
    heights = new_lst
    
    x = range(len(heights))
    plt.bar(np.arange(len(water)) - 0.5, water, color="darkblue", width=width)
    plt.bar(np.array(x) - 0.5, heights, color=color, width=width)
    plt.yticks(x)
    plt.xticks(x)
#     plt.grid()

def generate_random(a, b, size=10):
    return np.random.randint(a, b + 1, size=size)

def brute_force(heights: list[int]) -> int:
    area = 0
    n = len(heights)
    for x1 in range(n):
        for x2 in range(x1 + 1, n):
            width = x2 - x1
            height = min(heights[x1], heights[x2])
            area = max(area, width * height)

    return area
```

```python
h1 = [1, 8, 6, 2, 5, 4, 8, 3, 7]
w1 = [1, 1, 8, *[7] * 9, 8, 7, 7, 7, 7]
generate_walls(h1, w1)
```

```python
brute_force(h1)
```
