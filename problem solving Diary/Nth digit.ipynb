```python
import numpy as np
import itertools
```

```python
def nthdigit(n: int) -> int:
    for i in itertools.count(1):
        for j in str(i):
            n -= 1
            if n == 0:
                return int(j)

def nthpair(n: int) -> None:
    sno = 1
    for i in range(1, n + 1):
        snos = ()
        for j in str(i):
            snos += (sno,)
            sno += 1
        print(f"{i:>3}  {snos}")
```
