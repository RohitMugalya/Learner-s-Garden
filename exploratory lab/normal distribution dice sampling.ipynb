```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("dark_background")
```

```python
dice = [1, 2, 3, 4, 5, 6]
n = len(dice)

weights = [1.5, 1, 0.5, 0.5, 1, 1.5]
weights = [weight / n for weight in weights]
weights = [0.5, 0.25, 0.125, 0.0625, 0.03125, 0.03125]

sns.barplot(x=dice, y=weights, color="darkcyan")
```

```python
samples = 100000
n_dice = 3
observations = np.random.choice(dice, size=(samples, n_dice), p=weights)

outcomes = observations.sum(axis=1)

print(outcomes.mean())
sns.displot(outcomes, bins=18 - 2)
```
