```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

plt.style.use("dark_background")
```

```python
def scatter3d(X, color="blue", marker="o"):
    if X.shape[1] < 3:
        raise ValueError(f"{X.shape}")
    
    columns = X.columns[:3]
    xlabel, ylabel, zlabel = columns[0], columns[1], columns[2]

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(X[xlabel], X[ylabel], X[zlabel], c=color, marker=marker)
    ax.grid(False)

    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_zlabel(zlabel)


def f(x, y):
    return 5 * x + 3 * y + 2
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = f(x, y)

df = pd.DataFrame({"x": x, "y": y, "z": z})
# df
```

```python
scatter3d(df, color="black", marker='x')
```

```python
features = ["x", "y"]
label = "z"

X, y = df[features], df[label]
model = LinearRegression()
model.fit(X, y)

print("The accuracy:", model.score(X, y))
print("The MSE:", mean_squared_error(y, model.predict(X)))
```
