```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()

trigonometry = [
    [np.sin, np.cos, np.tan],
    [np.sinh, np.cosh, np.tanh],
    [np.sqrt, np.square, lambda arr: arr ** 2],
]

for i, row in enumerate(axes):
    for j, col in enumerate(row):
        col.plot(x, trigonometry[i][j](x))
fig, axes
```
