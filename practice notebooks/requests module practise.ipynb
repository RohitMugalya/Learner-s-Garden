```python
import requests 
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
r = requests.get(r"https://news.ycombinator.com/news", timeout=5)
```

```python
soup = BeautifulSoup(r.text)
print(soup.prettify())
```

```python
for child in soup.table.children:
    print(child.get_text())
```
